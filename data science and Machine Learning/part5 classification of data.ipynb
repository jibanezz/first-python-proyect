```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random data for four classes
np.random.seed(0)
# Class 1
class1_x = np.random.normal(2, 1, 100)
class1_y = np.random.normal(2, 1, 100)

# Class 2
class2_x = np.random.normal(6, 1, 100)
class2_y = np.random.normal(6, 1, 100)

# Class 3
class3_x = np.random.normal(2, 1, 100)
class3_y = np.random.normal(6, 1, 100)

# Class 4
class4_x = np.random.normal(6, 1, 100)
class4_y = np.random.normal(2, 1, 100)

# Combining data
data = np.column_stack((
    np.concatenate((class1_x, class2_x, class3_x, class4_x)),
    np.concatenate((class1_y, class2_y, class3_y, class4_y))
))

# Plotting the data
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], color='b', alpha=0.5)
plt.title('Clustering Dataset with 4 Classes')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python

# Perform K-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)
labels = kmeans.predict(data)
centroids = kmeans.cluster_centers_

# Plotting the clustered data
plt.figure(figsize=(8, 6))

# Plot data points
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.5, edgecolors='k')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='red', label='Centroids')

plt.title('K-means Clustering with 4 Clusters')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
