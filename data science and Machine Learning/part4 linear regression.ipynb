#### Regression is a model which can predict one value based on another value
###### for example, if we have a data about size of house and price of house we can predict unknown value

```python
size = [100,101,95,98,105]
price = [1005,1100,900,950,1200]
```

```python
# can you plot size vs price as figure
import matplotlib.pyplot as plt

plt.scatter(size,price)
plt.xlabel('size')
plt.ylabel('price')
plt.title('size vs price')
plt.show()
```
